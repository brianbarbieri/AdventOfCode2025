```python
data = """123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  """.split("\n")
with open("data.txt") as f:
  data = f.read().split("\n")
nums = [[int(elem) for elem in line.split(" ") if elem] for line  in data[:-1]] 
ops = [elem for elem in data[-1] if elem and elem != " "]
```

Star 1

```python
import numpy as np

rot_array = np.rot90(np.array(nums))
rot_ops = list(reversed(ops))

total = np.int64(0)  # avoid overflow
for i in range(len(rot_ops)):
    if rot_ops[i] == "*":
        total += np.prod(rot_array[i], dtype=np.int64)
    else:
        total += np.sum(rot_array[i], dtype=np.int64)

total
```

```python
data = """123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  """
with open("data.txt") as f:
  data = f.read()
head, sep, tail = data.rpartition('\n')
ops = [elem for elem in tail if elem and elem != " "]
lines = head.splitlines()

# Pad lines to the same length
max_len = max(len(line) for line in lines)
lines = [line.ljust(max_len) for line in lines]

# Rotate 90Â° clockwise while preserving spacing
rotated = ["".join(row[::-1]) for row in zip(*lines)]
rotated = ["".join(row[::-1]) for row in zip(*rotated)]
rotated = ["".join(row[::-1]) for row in zip(*rotated)]

total = np.int64(0)
arr = []
rot_ops = list(reversed(ops))
for val in rotated:
    parsed_val = val.replace(" ", "")
    if parsed_val:
        arr.append(int(parsed_val))
    else:
        op = rot_ops.pop(0)
        if op == "*":
            total += np.prod(np.array(arr), dtype=np.int64)
        else:
            total += np.sum(np.array(arr), dtype=np.int64)
        arr = []
op = rot_ops.pop(0)
if op == "*":
    total += np.prod(np.array(arr), dtype=np.int64)
else:
    total += np.sum(np.array(arr), dtype=np.int64)

total
```
