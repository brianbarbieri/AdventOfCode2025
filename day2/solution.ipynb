```python
total = 0
for rec in data:
    num1, num2 = rec.split("-")
    for i in range(int(num1), int(num2)+1):
        str_num = str(i)
        mid = len(str_num) // 2
        first_half = str_num[:mid]
        second_half = str_num[mid:]
        if first_half == second_half:
            total += i
total
```

```python
def is_invalid(s):
    L = len(s)
    # r = number of repeats (must be >= 2)
    for r in range(2, L + 1):
        if L % r != 0:
            continue
        size = L // r
        part = s[:size]
        if part * r == s:
            return True
    return False


total = 0

for rec in data:
    a, b = map(int, rec.split("-"))
    for i in range(a, b + 1):
        if is_invalid(str(i)):
            total += i

total
```
