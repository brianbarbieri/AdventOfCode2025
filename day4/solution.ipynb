```python
data = """..@@.@@@@.
@@@.@.@.@@
@@@@@.@.@@
@.@@@@..@.
@@.@@@@.@@
.@@@@@@@.@
.@.@.@.@@@
@.@@@.@@@@
.@@@@@@@@.
@.@.@@@.@.""".split("\n")
# with open("data.txt") as f:
#   data = f.read().split("\n")
```

Star 1

```python
rows, cols = len(data), len(data[0])

neighbors = [
    [data[r + dr][c + dc]
     for dr in (-1, 0, 1)
     for dc in (-1, 0, 1)
     if (dr != 0 or dc != 0)
     and 0 <= r + dr < rows
     and 0 <= c + dc < cols]
    for r in range(rows)
    for c in range(cols)
    if data[r][c] == "@"
]

sum(1 for cell in neighbors if cell.count("@") < 4)
```

Star 2

```python

def replace_char(string: str, index: int, char: str) -> str:
    return string[:index] + char + string[index+1:]


init = [row[:] for row in data]
clone = []
while "".join(init) != "".join(clone):
    clone = [row[:] for row in init]

    for r in range(rows):
        for c in range(cols):
            if clone[r][c] != "@":
                continue

            cell_neighbors = []
            for dr in (-1, 0, 1):
                for dc in (-1, 0, 1):
                    if dr == 0 and dc == 0:
                        continue

                    nr, nc = r + dr, c + dc
                    if 0 <= nr < rows and 0 <= nc < cols:
                        cell_neighbors.append(clone[nr][nc])

            if cell_neighbors.count("@") < 4:
                init[r] = replace_char(init[r], c, "x")
    # print("Step -----")
    # for row in init:
    #     print(row)
    # print("Total roles:", "".join(clone).count("x"))
"".join(clone).count("x")
```
