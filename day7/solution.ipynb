```python
total_breaks = 0
for i in range(len(data)):
    if i+1 >= len(data):
        break
    next_line = [char for char in data[i+1]]
    for j in range(len(data[i])):
        if (data[i][j] == "S" or data[i][j] == "|") and i+1 < len(data):
            if next_line[j] == ".":
                next_line[j] = "|"
            elif next_line[j] == "^":
                if data[i][j] == "|":
                    total_breaks += 1
                if j-1 != -1:
                    next_line[j-1] = "|"
                if j+1 < len(data[i]):
                    next_line[j+1] = "|"
    data[i+1] = "".join(next_line)

total_breaks
```

```python
from collections import defaultdict

data = """.......S.......
...............
.......^.......
...............
......^.^......
...............
.....^.^.^.....
...............
....^.^...^....
...............
...^.^...^.^...
...............
..^...^.....^..
...............
.^.^.^.^.^...^.
...............""".split("\n")
with open("data.txt") as f:
  data = f.read().split("\n")

beam_counts = defaultdict(int)

# initialize first beam(s)
for i, c in enumerate(data[0]):
    if c == "S":
        beam_counts[(1, i)] = 1  # start at row 1

max_row = len(data)
max_col = len(data[0])

# iterate row by row
for row in range(1, max_row):
    new_counts = defaultdict(int)
    for (r, c), count in beam_counts.items():
        if r != row:
            continue  # only propagate current row

        if r + 1 >= max_row:
            # reached bottom, keep count here
            new_counts[(r, c)] += count
            continue

        cell_below = data[r+1][c]
        if cell_below == ".":
            new_counts[(r+1, c)] += count
        elif cell_below == "^":
            if c - 1 >= 0:
                new_counts[(r+1, c-1)] += count
            if c + 1 < max_col:
                new_counts[(r+1, c+1)] += count

    beam_counts = new_counts

# total number of unique beam paths reaching the bottom row
total_paths = sum(beam_counts.values())
print(total_paths)  # 40
```
