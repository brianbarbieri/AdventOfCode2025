```python
# data = """3-5
# 10-14
# 16-20
# 12-18

# 1
# 5
# 8
# 11
# 17
# 32""".split("\n")
with open("data.txt") as f:
  data = f.read().split("\n")
ranges, ingredients = [], [] 
for item in data:
    if item == "":
        continue
    elif "-" in item:
        ranges.append(item)
    else:
        ingredients.append(item)
```

Star 1

```python
%%timeit
ranges, ingredients
range_pairs = []
for _range in ranges:
    min_val, max_val = map(int, _range.split("-"))
    range_pairs.append([min_val, max_val])

total = 0
for ing in ingredients:
    for min_val, max_val in range_pairs:
        if int(ing) >= min_val and int(ing) <= max_val:
            total += 1
            break

total
```

Star 2

```python
%%timeit
range_pairs = []
for _range in ranges:
    min_val, max_val = map(int, _range.split("-"))
    range_pairs.append([min_val, max_val])

range_pairs.sort(key=lambda x: x[0])
merged = []
for start, end in range_pairs:
    if not merged or start > merged[-1][1]:
        merged.append([start, end])
    else:
        merged[-1][1] = max(merged[-1][1], end)

sum([max_val + 1 - min_val for min_val, max_val in merged])
```
