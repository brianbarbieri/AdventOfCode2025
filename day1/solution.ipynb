```python
dial = 50
for rec in data:
    dir, num = rec[0], int(rec[1:])
    print(dir, num)
    if dir == "L":
        if dial - num < 0:
            rest = dial - num
            dial = 100 - rest
        else:
            dial -= num
    else:
        if dial + num > 99:
            dial = dial - num
        else:
            dial += num
    print(dial)
```
