```python
# data = """L68
# L30
# R48
# L5
# R60
# L55
# L1
# L99
# R14
# L82""".splitlines()
with open("data.txt") as f:
  data = f.readlines()
```

Star 1

```python
count, dial = 0, 50

for rec in data:
    d = rec[0]
    num = int(rec[1:])

    if d == "L":
        dial = (dial - num) % 100
    else:  # "R"
        dial = (dial + num) % 100

    if dial == 0:
        count += 1

print(count)
```

Star 2

```python
count = 0
dial = 50

for rec in data:
    d = rec[0]
    steps = int(rec[1:])

    k0 = ((-dial) % 100) if d == "R" else (dial % 100)

    if k0 == 0:
        hits = steps // 100
    else:
        hits = 1 + (steps - k0) // 100 if steps >= k0 else 0

    count += hits
    dial = (dial + steps) % 100 if d == "R" else (dial - steps) % 100

print(count)
```
