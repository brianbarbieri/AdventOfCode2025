Star 1

```python
def highest_num(s):
    max_val = 0
    for i in range(len(s)):
        for j in range(i + 1, len(s)):
            val = int(s[i] + s[j])
            if val > max_val:
                max_val = val
    return int(max_val)


total = sum(highest_num(rec) for rec in data)
print(total)
```

```python
def max_subsequence(s, k=12):
    stack = []
    to_remove = len(s) - k

    for ch in s:
        while stack and to_remove and stack[-1] < ch:
            stack.pop()
            to_remove -= 1
        stack.append(ch)

    return int("".join(stack[:k]))


total = sum(max_subsequence(rec, 12) for rec in data)
print(total)
```
